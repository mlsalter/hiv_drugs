```python
#Dependencies
import pandas as pd
import glob
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

```python
case_files = glob.glob("Resources/*.csv")
case_files
```

```python
#read in csv's to use
diag_nat_cr = pd.read_csv('Resources/Diag_National_CandR.csv')
diag_st_cr = pd.read_csv('Resources/Diag_State_CandR.csv')
diag_cnty_cr = pd.read_csv('Resources/Diag_County_CandR.csv')

diag_nat_cr['Geography'] = 'US'
diag_nat_cr
```

```python
diag_nat_cr.columns
# diag_st_cr.columns
# diag_cnty_cr.columns
```

```python
#Rename column headers
#Rate is per 100,000
diag_nat_cr = diag_nat_cr.rename(columns= {'2008':'Cases_08','Unnamed: 1':'Rate_08','2009':'Cases_09','Unnamed: 3':'Rate_09',
                                           '2010':'Cases_10','Unnamed: 5':'Rate_10','2011':'Cases_11','Unnamed: 7':'Rate_11',
                                           '2012':'Cases_12','Unnamed: 9':'Rate_12','2013':'Cases_13','Unnamed: 11':'Rate_13',
                                           '2014':'Cases_14','Unnamed: 13':'Rate_14','2015':'Cases_15','Unnamed: 15':'Rate_15',
                                           '2016':'Cases_16','Unnamed: 17':'Rate_16','2017':'Cases_17','Unnamed: 19':'Rate_17',
                                           '2018':'Cases_18','Unnamed: 21':'Rate_18'})
diag_st_cr = diag_st_cr.rename(columns={'2008':'Cases_08','Unnamed: 2':'Rate_08','2009':'Cases_09','Unnamed: 4':'Rate_09',
                                           '2010':'Cases_10','Unnamed: 6':'Rate_10','2011':'Cases_11','Unnamed: 8':'Rate_11',
                                           '2012':'Cases_12','Unnamed: 10':'Rate_12','2013':'Cases_13','Unnamed: 12':'Rate_13',
                                           '2014':'Cases_14','Unnamed: 14':'Rate_14','2015':'Cases_15','Unnamed: 16':'Rate_15',
                                           '2016':'Cases_16','Unnamed: 18':'Rate_16','2017':'Cases_17','Unnamed: 20':'Rate_17',
                                           '2018':'Cases_18','Unnamed: 22':'Rate_18'})
diag_cnty_cr = diag_cnty_cr.rename(columns={'2008':'Cases_08','Unnamed: 2':'Rate_08','2009':'Cases_09','Unnamed: 4':'Rate_09',
                                           '2010':'Cases_10','Unnamed: 6':'Rate_10','2011':'Cases_11','Unnamed: 8':'Rate_11',
                                           '2012':'Cases_12','Unnamed: 10':'Rate_12','2013':'Cases_13','Unnamed: 12':'Rate_13',
                                           '2014':'Cases_14','Unnamed: 14':'Rate_14','2015':'Cases_15','Unnamed: 16':'Rate_15',
                                           '2016':'Cases_16','Unnamed: 18':'Rate_16','2017':'Cases_17','Unnamed: 20':'Rate_17',
                                           '2018':'Cases_18','Unnamed: 22':'Rate_18'})
#Put 'Geography' column in the right spot
diag_nat_cr = diag_nat_cr[['Geography','Cases_08','Rate_08','Cases_09','Rate_09','Cases_10','Rate_10',
                           'Cases_11','Rate_11','Cases_12','Rate_12','Cases_13','Rate_13','Cases_14','Rate_14',
                           'Cases_15','Rate_15','Cases_16','Rate_16','Cases_17','Rate_17','Cases_18','Rate_18']]
diag_nat_cr
```

```python
#Remove unwanted row
diag_nat_df = diag_nat_cr.iloc[1:]
diag_st_df = diag_st_cr.iloc[1:]
diag_cnty_df = diag_cnty_cr.iloc[1:]
diag_nat_df
```

```python
diag_cnty_df.head()
```

```python
def parsed(x):
    try:
        return int(x)
    except:
        return x

diag_all_df = pd.concat([diag_nat_df,diag_st_df,diag_cnty_df])
diag_all_df = diag_all_df.replace({'Data suppressed': np.nan, ',': ''}, regex=True)
diag_all_df = diag_all_df.apply(lambda x: parsed(x))
diag_all_df.head()
```

```python
diag_all_df.to_csv('overall_diagnosis.csv',index=False)
```

```python
#Use skipna to skip the null values so as not to mess with our numbers, but should be pointed out there are some
diag_all = diag_all_df
diag_all['Avg_Cases_Before'] = diag_all[['Cases_08','Cases_09', 'Cases_10', 'Cases_11','Cases_12']].mean(axis=1,skipna=True).round(0)
diag_all['Avg_Rate_Before'] = diag_all[['Rate_08','Rate_09', 'Rate_10', 'Rate_11','Rate_12']].mean(axis=1,skipna=True).round(1)
diag_all['Avg_Cases_After'] = diag_all[['Cases_13','Cases_14', 'Cases_15', 'Cases_16','Cases_17','Cases_18']].mean(axis=1,skipna=True).round(0)
diag_all['Avg_Rate_After'] = diag_all[['Rate_13','Rate_14', 'Rate_15', 'Rate_16','Rate_17','Rate_18']].mean(axis=1,skipna=True).round(1)
diag_all_df.head()
```

```python
diag_all_avg = diag_all[['Geography','Avg_Cases_Before','Avg_Rate_Before','Avg_Cases_After','Avg_Rate_After']]
diag_all_avg.head()
```

```python
diag_all_avg.to_csv('overall_diag_ba_avg.csv',index=False)
```

```python
#Find the difference between Before and After
diag_all_diff =  diag_all_avg
diag_all_diff['Difference_Cases'] = diag_all_diff['Avg_Cases_Before'] - diag_all_diff['Avg_Cases_After']
diag_all_diff['Difference_Rates'] = diag_all_diff['Avg_Rate_Before'] - diag_all_diff['Avg_Rate_After']
diag_all_diff
```

```python
diag_all_diff.to_csv('overall_diag_ba_diff.csv')
```
