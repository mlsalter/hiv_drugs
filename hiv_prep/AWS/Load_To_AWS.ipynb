```python
# SQL ALchemy 
from sqlalchemy import create_engine

# Import PyMySQL
import pymysql
pymysql.install_as_MySQLdb()

# Config Variables 
from config import remote_db_endpoint, remote_db_port
from config import remote_gwsis_dbname, remote_gwsis_dbuser, remote_gwsis_dbpwd
# from config import local_db_user, local_db_pwd, local_db_endpoint, local_db_port, local_db_name

# Import Pandas 
import pandas as pd
```

```python
# Create engine and pass mySQL connnection 

# CSV Files
diag_data = '../HIV_Data/Output/master_diag.csv'
diag_df = pd.read_csv(diag_data)


# Cloud MySQl database connection on AWS
engine = create_engine(f"mysql://{remote_gwsis_dbuser}:{remote_gwsis_dbpwd}@{remote_db_endpoint}:{remote_db_port}/{remote_gwsis_dbname}")
```

```python
# Create remote database engine connection 
conn = engine.connect()
```

```python
# Dataframe to SQL
diag_df.to_sql(name='diag_data', if_exists='append', con=conn, chunksize=500, index=False)

# death_df.to_sql(name='death_data', if_exists='append', con=conn, chunksize=500, index=False)
```

```python
remote_diag_data = pd.read_sql("SELECT * FROM diag_data", conn)
print(len(remote_diag_data))
remote_diag_data.head()
```
