```python
#Dependencies
import pandas as pd
import glob
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

```python
#Check all the csv's
death_files = glob.glob("Resources/*.csv")
death_files
```

```python
#Read in the csv's
death_nat = pd.read_csv('Resources/Deaths_National_CandR.csv') 
death_st = pd.read_csv('Resources/Deaths_State_CandR.csv')
death_nat.head()
```

```python
#Rename column headers of the national
death_nat = death_nat.rename(columns= {'2008':'Deaths_08','Unnamed: 2':'Death_Rate_08','2009':'Deaths_09','Unnamed: 4':'Death_Rate_09',
                                           '2010':'Deaths_10','Unnamed: 6':'Death_Rate_10','2011':'Deaths_11','Unnamed: 8':'Death_Rate_11',
                                           '2012':'Deaths_12','Unnamed: 10':'Death_Rate_12','2013':'Deaths_13','Unnamed: 12':'Death_Rate_13',
                                           '2014':'Deaths_14','Unnamed: 14':'Death_Rate_14','2015':'Deaths_15','Unnamed: 16':'Death_Rate_15',
                                           '2016':'Deaths_16','Unnamed: 18':'Death_Rate_16','2017':'Deaths_17','Unnamed: 20':'Death_Rate_17'})
death_nat.head()
```

```python
#Remove unwanted row , add new 'Type' column, and rename 'Geography'
death_nat_df = death_nat.iloc[1:].rename(columns={'Geography':'Name'})
death_nat_df['Type'] = 'National'
death_nat_df
```

```python
#Pivot tabel
death_rate_piv = death_st.pivot(index='Geography',columns='Year',values='Rate per 100000')
death_rate_piv = death_rate_piv.reset_index()
death_rate_piv.head()
```

```python
death_rate_rename = death_rate_piv.rename(columns= {'Geography':'Name',2008:'Death_Rate_08',2009:'Death_Rate_09',2010:'Death_Rate_10',2011:'Death_Rate_11',
                                            2012:'Death_Rate_12',2013:'Death_Rate_13',2014:'Death_Rate_14',2015:'Death_Rate_15',
                                           2016:'Death_Rate_16',2017:'Death_Rate_17'})
death_rate_rename.head()
```

```python
death_cases_piv = death_st.pivot(index='Geography',columns='Year',values='Cases')
death_cases_piv = death_cases_piv.reset_index()
death_cases_piv.head()
```

```python
death_cases_rename = death_cases_piv.rename(columns= {'Geography':'Name',2008:'Deaths_08',2009:'Deaths_09',2010:'Deaths_10',2011:'Deaths_11',
                2012:'Deaths_12',2013:'Deaths_13',2014:'Deaths_14',2015:'Deaths_15',
                2016:'Deaths_16',2017:'Deaths_17'})
death_cases_rename.head()
```

```python
death_st_df = pd.merge(death_cases_rename,death_rate_rename,how='outer',on='Name')
#Place in the same order as national so to concat
death_st_df = death_st_df[['Name','Deaths_08','Death_Rate_08','Deaths_09','Death_Rate_09','Deaths_10','Death_Rate_10','Deaths_11','Death_Rate_11',
                                   'Deaths_12','Death_Rate_12','Deaths_13','Death_Rate_13','Deaths_14','Death_Rate_14','Deaths_15','Death_Rate_15',
                                   'Deaths_16','Death_Rate_16','Deaths_17','Death_Rate_17']]
death_st_df['Type'] = 'State'
death_st_df.head()
```

```python
deaths_df = pd.concat([death_nat_df,death_st_df])
deaths_df = deaths_df[['Name','Type','Deaths_08','Death_Rate_08','Deaths_09','Death_Rate_09','Deaths_10','Death_Rate_10','Deaths_11','Death_Rate_11',
                      'Deaths_12','Death_Rate_12','Deaths_13','Death_Rate_13','Deaths_14','Death_Rate_14','Deaths_15','Death_Rate_15',
                      'Deaths_16','Death_Rate_16','Deaths_17','Death_Rate_17']]
deaths_df.head()
```
