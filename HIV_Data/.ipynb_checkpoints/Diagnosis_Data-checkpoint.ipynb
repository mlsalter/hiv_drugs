```python
diag_all_files = glob.glob("Resources/*All*.csv")
diag_all_files
```

```python
#Check data is reading in and how it looks
diag_test = pd.read_csv('Resources/Diag_All_National_CandR.csv')
# diag_test = pd.read_csv('Resources/Diag_All_State_CandR.csv')
# diag_test = pd.read_csv('Resources/Diag_All_County_CandR.csv')

diag_test.head()
# diag_test.columns
```

```python
def parsed(x):
    try:
        return int(x)
    except:
        return x
i=0

for f in diag_all_files:
    diag_all_df = pd.read_csv(f)
    #Shorten and rename columns (for usability) American Indian/Alaska Native=AI_AN, Asian, Black/African American=Black, Hispanic/Latino=Latino, Multiple races=Multi, Native Hawaiian/Other Pacific Islander=NH_OPI, White
    diag_all_df = diag_all_df.rename(columns={'Geography':'Full Name','2008':'Diag 08','Unnamed: 2':'Diag Rate 08','2009':'Diag 09','Unnamed: 4':'Diag Rate 09',
                                           '2010':'Diag 10','Unnamed: 6':'Diag Rate 10','2011':'Diag 11','Unnamed: 8':'Diag Rate 11',
                                           '2012':'Diag 12','Unnamed: 10':'Diag Rate 12','2013':'Diag 13','Unnamed: 12':'Diag Rate 13',
                                           '2014':'Diag 14','Unnamed: 14':'Diag Rate 14','2015':'Diag 15','Unnamed: 16':'Diag Rate 15',
                                           '2016':'Diag 16','Unnamed: 18':'Diag Rate 16','2017':'Diag 17','Unnamed: 20':'Diag Rate 17',
                                           '2018':'Diag 18','Unnamed: 22':'Diag Rate 18'
    })

    diag_all_df = diag_all_df.iloc[1:].replace({'Data suppressed': np.nan, ',': ''}, regex=True)
    diag_all_df = diag_all_df.apply(lambda x: parsed(x))
    if i>1:
        diag_all_df1 = diag_all_df
    elif i>0:
        diag_all_df2 = diag_all_df
    else:
        diag_all_df0 = diag_all_df
    i = i + 1

# Concat all the df's together into one
diag_all_data = pd.concat([diag_all_df0,diag_all_df1,diag_all_df2]).reset_index(drop=True)
diag_all_data.head()
```
