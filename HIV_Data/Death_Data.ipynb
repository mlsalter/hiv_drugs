```python
#Dependencies
import pandas as pd
import glob
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

```python
death_all_files = glob.glob("Resources/Death_All*.csv")
death_all_files
```

```python
#Check data is reading in and how it looks
death_test = pd.read_csv('Resources/Death_All_National.csv')
# death_test = pd.read_csv('Resources/Death_All_State')

death_test.head()
# ddeath_test.columns
```

```python
def parsed(x):
    try:
        return int(x)
    except:
        return x
i=0

for f in death_all_files:
    death_all_df = pd.read_csv(f)
    #Shorten and rename columns
    death_all_df = death_all_df.rename(columns={'Year':'Year','Geography':'Full Name', 'Cases':'All Deaths','Rate per 100000':'All Death Rates'
    })

    death_all_df = death_all_df.replace({'Data suppressed': np.nan, ',': ''}, regex=True)
    death_all_df = death_all_df.apply(lambda x: parsed(x))
    death_all_df['Year'] = death_all_df['Year'].apply(lambda x: parsed(x))
    if i>0:
        death_all_df0 = death_all_df
    else:
        death_all_df1 = death_all_df
    i = i + 1

# Concat all the df's together into one
death_all_data = pd.concat([death_all_df0,death_all_df1]).reset_index(drop=True)
death_all_data.head()
```

```python
death_all_data.to_csv('Output/death_all.csv',index=False)
```
