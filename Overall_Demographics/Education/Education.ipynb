```python
import pandas as pd
import numpy as np
import glob
```

```python
edu_files = glob.glob_csv('Resources/*.csv')
print(edu_files)
```

```python
edu_file_08 = pd.read_csv('Resources/ACS_08_1YR_S1501_with_ann.csv')
edu_file_08.head()
```

```python
new_header = edu_file_08.iloc[0]
edu_file_08 = edu_file_08[1:] 
edu_file_08.columns = new_header 
ed_df_08 = edu_file_08
ed_df_08.head()
```

```python
edu_totals_08 = ed_df_08.loc[:, (ed_df_08.columns.str.startswith('Total; Estimate; Population'))|
                            (ed_df_08.columns.str.startswith('Geography'))]
edu_totals_08.head()
```

```python
#Read in the files by looping through files and then appending to  a list
edu_dfs = [] 

for file in edu_files:
    edu_dfs.append(pd.read_csv(file))
```

```python
#Concatanate income_dfs to one df
concat_edu_df = pd.concat(edu_dfs, ignore_index=False)

concat_edu_df.head()
```
